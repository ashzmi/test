# Analyze User Behavior & Transactions

### **Merge Dataframe**

```python
behavior_df = (
    events_df
    .merge(session_df, on="session_id", how="left", suffixes=("_event", "_session"))
    .merge(customers_df, on="customer_id", how="left")
    .merge(products_df, on="product_id", how="left")
)
```

```python
behavior_df.head()
```

```python
order_summary = (
    orders_df.groupby("customer_id", as_index=False)
    .agg(
        total_orders=("order_id", "nunique"),
        total_spent=("total_usd", "sum"),
        last_order_time=("order_time", "max")
    )
)
```

```python
order_summary
```

```python
behavior_transactions_df = behavior_df.merge(order_summary, on="customer_id", how="left")
```

```python
behavior_transactions_df.head(20)
```

```python
# Validate key relationship
behavior_transactions_df.groupby('session_id')['customer_id'].nunique().max()
```

show the behaviour of customer for example 12360

```python
cust_id = 12360

cust_df = user_Behaviour_df[user_Behaviour_df['customer_id'] == cust_id]

sessions = cust_df['session_id'].unique()
session_count = cust_df['session_id'].nunique()

print("Customer ID:", cust_id)
print("Total Sessions:", session_count)
print("Session IDs:", sessions)
```

```python
user_Behaviour_df[user_Behaviour_df['session_id'] == 19187]
```

*From the example of analysis above a customer can have multiple sessions, and each session consists of multiple events. Each event represents an action taken by the customer, such as page view, add to cart, checkout or purchase. However, not every session completes all steps of the customer journey.*

```python
# Understand User Activity (Basic EDA)
# Track user journey
user_Behaviour_df['event_type'].value_counts()
```

```python
event_counts = user_Behaviour_df['event_type'].value_counts()

total = event_counts['page_view']

add_to_cart_rate = event_counts['add_to_cart'] / total
checkout_rate = event_counts['checkout'] / total
purchase_rate = event_counts['purchase'] / total

print (f"Total Page View:  {total}")
print(f"Add to Cart Rate: {add_to_cart_rate:.2%}")
print(f"Checkout Rate: {checkout_rate:.2%}")
print(f"Purchase Rate: {purchase_rate:.2%}")
```

* Most user activities are page views (393k), but only a small portion convert into purchases (24k)
- While 26.54% of users add items to cart, only 6.23% complete a purchase, indicating a significant drop-off throughout the conversion funnel.*
