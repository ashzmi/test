```python
# Convert date/time columns into datetime format
# This is necessary for time-based feature engineering such as session duration and account age
customers["signup_date"] = pd.to_datetime(customers["signup_date"], errors="coerce")
sessions["start_time"] = pd.to_datetime(sessions["start_time"], errors="coerce")
events["timestamp"] = pd.to_datetime(events["timestamp"], errors="coerce")
```

Datetime conversion enables time-based analysis and feature engineering. For example, we later compute session duration and how long a customer has been registered before the session occurred.

```python
# Convert date/time columns into datetime format
# This is necessary for time-based feature engineering such as session duration and account age
customers = customers.drop_duplicates()
sessions = sessions.drop_duplicates()
events = events.drop_duplicates()
```

```python
events["event_type"] = events["event_type"].astype(str).str.strip().str.lower()
```

Standardizing event labels ensures that all events of the same type are grouped correctly during aggregation, avoiding duplication caused by inconsistent formatting such as capitalization and extra spaces.

```python
event_counts = events.pivot_table(
    index="session_id",
    columns="event_type",
    values="event_id",
    aggfunc="count",
    fill_value=0
).reset_index()
```

This step transforms event-level logs into session-level behavioral features. It allows us to quantify how much browsing, cart activity, and purchasing happened in each session.

```python
#just to make sure key event column exists
for col in ["page_view", "add_to_cart", "purchase"]:
    if col not in event_counts.columns:
        event_counts[col] = 0
```

```python
# These summarize how active and engaged each session was
session_agg = events.groupby("session_id").agg(
    num_events=("event_id", "count"),
    num_unique_products=("product_id", "nunique"),
    total_qty_added=("qty", "sum"),
    max_cart_size=("cart_size", "max"),
    avg_discount_pct=("discount_pct", "mean"),
    total_amount_usd=("amount_usd", "sum"),
    first_event_time=("timestamp", "min"),
    last_event_time=("timestamp", "max")
).reset_index()
```

This block generates higher-level behavioral features for each session. These features help the model distinguish between light browsing, engaged shopping, and sessions that are closer to conversion.

```python
# Fill missing aggregated numeric fields with 0
#cleaning the Nan to 0
numeric_cols = ["num_unique_products", "total_qty_added", "max_cart_size", "avg_discount_pct", "total_amount_usd"]
session_agg[numeric_cols] = session_agg[numeric_cols].fillna(0)
```

```python
session_agg["session_duration_min"] = (
    (session_agg["last_event_time"] - session_agg["first_event_time"]).dt.total_seconds() / 60
)

session_agg["session_duration_min"] = session_agg["session_duration_min"].fillna(0)
```

add new column to the table

Session duration was computed as the time difference between the first and last event within each session. This feature captures user engagement level, as longer sessions may indicate higher interest or decision-making activity.

```python
# Merge and Combine event counts and aggregated behavioral metrics into one session-level feature table
session_features = event_counts.merge(session_agg, on="session_id", how="left")
print(session_features.head())
```

```python
# Keep only sessions where at least one item was added to cart
# These are the sessions relevant for cart abandonment analysis
cart_sessions = session_features[session_features["add_to_cart"] > 0].copy()

# Define cart abandonment label using business logic
# abandoned = 1 if the session had add-to-cart activity but no purchase
# abandoned = 0 if the session had add-to-cart activity and completed a purchase
cart_sessions["abandoned"] = np.where(
    cart_sessions["purchase"] > 0,
    0,
    1
)

print(cart_sessions[["session_id", "add_to_cart", "purchase", "abandoned"]].head())
print(cart_sessions["abandoned"].value_counts())
```

This is a form of target engineering where cart abandonment is defined using business logic. A session is labeled as abandoned if items were added to the cart but no purchase was completed.

Numeric and categorical features were separated to ensure appropriate preprocessing. Numerical variables are directly usable by models, while categorical variables require encoding before being used in machine learning algorithms.

```python
numeric_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="median"))
])

categorical_transformer = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy="most_frequent")),
    ("onehot", OneHotEncoder(handle_unknown="ignore"))
])

preprocessor = ColumnTransformer(transformers=[
    ("num", numeric_transformer, numeric_features),
    ("cat", categorical_transformer, categorical_features)
])
```

Prepares data for modeling:

- numeric nulls → filled with median
- categorical nulls → filled with most frequent value
- categories → converted into machine-readable columns using one-hot encoding

This preprocessing pipeline handles missing values and encodes categorical variables into numerical format using imputation and one-hot encoding, ensuring the dataset is fully compatible with machine learning models.

```python
# Train the model on historical session data
# The model learns patterns associated with abandoned vs completed cart sessions
model.fit(X_train, y_train)
```

During training, the model learns statistical patterns from historical sessions, such as whether shorter sessions, certain traffic sources, or certain device types are associated with higher abandonment risk.

# Product Recommender

```python
# ── Module 3: Product Recommender (fixed) ──
from itertools import combinations
from collections import defaultdict
```

Filter the right events

```python
# Use add_to_cart events since purchase rows have no product info
purchases = behavior_df[behavior_df["event_type"] == "add_to_cart"][
    ["customer_id", "product_id", "name_y", "category"]
].dropna()

print(f"Working with {len(purchases)} add-to-cart events")
print(f"Unique products: {purchases['product_id'].nunique()}")
print(f"Unique customers: {purchases['customer_id'].nunique()}")
```

```python
# Fix product_id display (remove decimal)
purchases["product_id"] = purchases["product_id"].astype(int)
product_lookup.index = product_lookup.index.astype(int)
```

Build the co-cart matrix

```python
# Rebuild co-cart matrix with clean IDs
co_purchases = defaultdict(int)

for customer, group in purchases.groupby("customer_id"):
    products = group["product_id"].tolist()
    if len(products) < 2:
        continue
    for p1, p2 in combinations(set(products), 2):
        co_purchases[(p1, p2)] += 1
        co_purchases[(p2, p1)] += 1

print(f"\nTotal product pairs found: {len(co_purchases)}")
```

```python
# ── Product lookup table (id → name & category) ──
product_lookup = (
    purchases[["product_id", "name_y", "category"]]
    .drop_duplicates("product_id")
    .set_index("product_id")
)
```

The recommend function

```python
# ── Recommend function ──
def recommend(product_id, top_n=5):
    scores = {
        p2: count
        for (p1, p2), count in co_purchases.items()
        if p1 == product_id
    }
    if not scores:
        print(f"No recommendations found for product {product_id}")
        return pd.DataFrame()

    top = sorted(scores.items(), key=lambda x: x[1], reverse=True)[:top_n]

    result = []
    for pid, score in top:
        if pid in product_lookup.index:
            row = product_lookup.loc[pid]
            result.append({
                "product_id": pid,
                "name": row["name_y"],
                "category": row["category"],
                "times_carted_together": score
            })

    return pd.DataFrame(result)
```

```python
# ── Test it ──
sample_product = purchases["product_id"].value_counts().index[0]
sample_name = product_lookup.loc[sample_product, "name_y"]

print(f"\nTop recommendations for: '{sample_name}'")
print(recommend(sample_product))
```

```python
# ── Search by product name (for demo) ──
def recommend_by_name(product_name_query, top_n=5):
    # Find matching products (case-insensitive partial match)
    matches = product_lookup[
        product_lookup["name_y"].str.contains(product_name_query, case=False, na=False)
    ]

    if matches.empty:
        print(f"No product found matching '{product_name_query}'")
        return pd.DataFrame()

    # Take the first match
    product_id = matches.index[0]
    product_name = matches.loc[product_id, "name_y"]
    print(f"Showing recommendations for: '{product_name}'\n")

    return recommend(product_id, top_n)


# ── Test with name search ──
print(recommend_by_name("Headphones"))
```

```python
# MODULE 3: PRODUCT RECOMMENDER — FINAL DEMO

def demo_recommender():
    print("=" * 55)
    print("   SMART E-COMMERCE — PRODUCT RECOMMENDER (Module 3)")
    print("=" * 55)

    # ── Dataset summary ──
    print("\n📦 DATASET SUMMARY")
    print(f"   Total cart events    : {len(purchases):,}")
    print(f"   Unique products      : {purchases['product_id'].nunique()}")
    print(f"   Unique customers     : {purchases['customer_id'].nunique()}")
    print(f"   Product pairs tracked: {len(co_purchases):,}")
    print(f"   Categories covered   : {purchases['category'].nunique()}")

    # ── Category breakdown ──
    print("\n🗂️  PRODUCTS PER CATEGORY")
    for cat, count in product_lookup["category"].value_counts().items():
        print(f"   {cat:<20} {count} products")

    # ── Live recommendations per category ──
    print("\n🎯 SAMPLE RECOMMENDATIONS BY CATEGORY")
    print("-" * 55)

    categories = product_lookup["category"].unique()

    for cat in categories:
        # Get most popular product in this category
        cat_products = purchases[purchases["category"] == cat]["product_id"]
        if cat_products.empty:
            continue

        pid = cat_products.value_counts().index[0]
        pname = product_lookup.loc[pid, "name_y"]
        recs = recommend(pid, top_n=3)

        print(f"\n  📌 Category : {cat}")
        print(f"  Input       : {pname}")
        print(f"  Recommended :")

        if recs.empty:
            print("     No recommendations found")
        else:
            for i, row in recs.iterrows():
                print(f"     {i+1}. {row['name']:<40} ({row['category']}) — score: {row['times_carted_together']}")

    print("\n" + "=" * 55)
    print("  ✅ Module 3 complete — Recommender is working!")
    print("=" * 55)

# ── Run the demo ──
demo_recommender()
```
